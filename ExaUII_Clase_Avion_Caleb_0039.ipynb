```python
# clase Avion, Atributos: aerolinea, capacidad, destino.
# Funciones: leer_datos() y mostrar_datos()
# crear 2 objetos ---> avion1, avion2
# Leer datos
class Avion:
    def __init__(self, aerolinea, capacidad, destino):
        self.aerolinea = aerolinea
        self.capacidad = capacidad
        self.destino = destino

    def leer_datos(self):
        self.aerolinea = input("Aerolinea: ")
        self.capacidad = int(input("Capacidad: "))
        self.destino = input("Destino: ")

    def mostrar_datos(self):
        print(f"Aerolinea: {self.aerolinea}")
        print(f"Capacidad: {self.capacidad}")
        print(f"Destino: {self.destino}")
        return self.aerolinea, self.capacidad, self.destino

print("==============================")
print("------- Avion numero 1 -------")
avion1 = Avion("", "", "")
avion1.leer_datos()
avion1.mostrar_datos()
print("==============================")
print("------- Avion numero 2 -------")
avion2 = Avion("", "", "")
avion2.leer_datos()
avion2.mostrar_datos()
print("==============================")
print("Software realizado por Caleb Dorado, NL = 14")
```
