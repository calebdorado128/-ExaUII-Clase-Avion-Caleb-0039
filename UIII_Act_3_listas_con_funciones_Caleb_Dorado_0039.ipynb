## Ejemplos de listas con funciones

```python
# Actividad: Listas con funciones - Tema: Panadero
# Programador: [Tu Nombre]

def Mi_Panaderia():
    print("\n--- BIENVENIDO A LA PANADERÍA 'EL TRIGO DE ORO' ---")
    panes = ["Bolillo", "Concha", "Dona", "Cuernito", "Telera"]

    def mostrar_panes():
        print(f"\nInventario actual de panes: {panes}")
        print(f"Total de variedades: {len(panes)}")

    def agregar_pan():
        nuevo_pan = input("Ingrese el nombre del nuevo pan a producir: ")
        panes.append(nuevo_pan)
        print(f"¡'{nuevo_pan}' ha sido horneado y agregado al mostrador!")

    def vender_pan():
        mostrar_panes()
        posicion = int(input("Ingrese el número de índice del pan a vender (empezando desde 0): "))
        if 0 <= posicion < len(panes):
            vendido = panes.pop(posicion)
            print(f"Se ha vendido: {vendido}. ¡Cliente satisfecho!")
        else:
            print("Ese lugar en la vitrina está vacío.")

    def modificar_pan():
        mostrar_panes()
        indice = int(input("¿Qué índice desea modificar? "))
        if 0 <= indice < len(panes):
            antiguo = panes[indice]
            nuevo = input(f"Cambiar '{antiguo}' por: ")
            panes[indice] = nuevo
            print(f"Se ha actualizado '{antiguo}' por '{nuevo}'.")
        else:
            print("Índice no válido.")

    def buscar_pan():
        busqueda = input("¿Qué pan busca el cliente? ")
        if busqueda in panes:
            print(f"¡Sí tenemos {busqueda} en existencia!")
        else:
            print(f"Lo sentimos, el {busqueda} se ha agotado.")

    # Menú interactivo
    while True:
        print("\n--- MENÚ DEL PANADERO ---")
        print("1. Ver inventario")
        print("2. Hornear nuevo pan (Agregar)")
        print("3. Vender pan (Eliminar)")
        print("4. Cambiar receta (Modificar)")
        print("5. Buscar pan")
        print("6. Cerrar panadería (Salir)")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            mostrar_panes()
        elif opcion == "2":
            agregar_pan()
        elif opcion == "3":
            vender_pan()
        elif opcion == "4":
            modificar_pan()
        elif opcion == "5":
            buscar_pan()
        elif opcion == "6":
            print("Cerrando cortinas... ¡Hasta mañana!")
            break
        else:
            print("Opción no válida, intente de nuevo.")

# Ejecutar la función principal
Mi_Panaderia()

print("Hecho por Caleb 0039")
```
